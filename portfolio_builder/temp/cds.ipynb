```python
import pandas as pd
import numpy as np

d = {"Time (years)": [1, 2, 3, 4, 5]}
df = pd.DataFrame(data=d)
upod = 0.03
uncond_survival_rate = 1 - upod
r = 0.050
R = 0.40
recovery_rate = 0.25


df["Unconditional Default Probability"] = upod
df["Unconditional Survival Probability"] = uncond_survival_rate
print("=" * 60)
print("Table 1. Unconditional default and survival probabilities")
print("=" * 60)
df
```

```python
df1 = df.copy()
df1["Conditional Default Probability"] = (
    df1["Unconditional Default Probability"]
).round(4)
df1["Conditional Survival Probability"] = 0.000
for i in range(0, len(df1["Time (years)"])):
    df1["Conditional Survival Probability"][i] = (
        (df1["Unconditional Survival Probability"][i]) ** (df1["Time (years)"][i])
    ).round(4)
for i in range(1, len(df["Time (years)"])):
    df1["Conditional Default Probability"][i] = (
        (df1["Unconditional Default Probability"][i])
        * (df1["Conditional Survival Probability"][i - 1])
    ).round(4)
df1.drop(
    ["Unconditional Default Probability", "Unconditional Survival Probability"],
    axis=1,
    inplace=True,
)
print("=" * 57)
print("Table 2. Conditional default and survival probabilities")
print("=" * 57)
df1
```

```python
df2 = df1.copy()
df2.drop(["Conditional Default Probability"], axis=1, inplace=True)
df2["Expected Payment"] = (df2["Conditional Survival Probability"]).round(4)
df2["Financial Discount Factor"] = (1 * np.exp(-r * df2["Time (years)"])).round(4)
df2["APV of Expected Payment"] = (
    df2["Expected Payment"] * df2["Financial Discount Factor"]
).round(4)
df2["Expected Payment"] = (df2["Expected Payment"]).astype(str) + "s"
df2.loc[len(df2.index)] = [
    "Total",
    "",
    "",
    "",
    np.sum(df2["APV of Expected Payment"]).round(4),
]
df2["APV of Expected Payment"] = df2["APV of Expected Payment"].astype(str) + "s"
print("=" * 73)
print("Table 3. Calculation of the actuarial present value of expected payments")
print("=" * 73)
df2
```

```python
df3 = df1.copy()
df3.drop(["Conditional Survival Probability"], axis=1, inplace=True)
df3["Time (years)"] = 0.00
for i in range(0, len(df["Time (years)"])):
    df3["Time (years)"][i] = df1["Time (years)"][i] - 0.5
df3["LGD"] = 1 - R
df3["Expected Payoff($)"] = (df3["Conditional Default Probability"] * df3["LGD"]).round(
    4
)
df3["Financial Discount Factor"] = (1 * np.exp(-r * df3["Time (years)"])).round(4)
df3["APV of Expected Payoff($)"] = (
    df3["Expected Payoff($)"] * df3["Financial Discount Factor"]
).round(4)
df3.loc[len(df3.index)] = [
    "Total",
    "",
    "",
    "",
    "",
    np.sum(df3["APV of Expected Payoff($)"]).round(4),
]
print("=" * 97)
print(
    "Table 4. Calculation of the actuarial present value of expected payoffs. Notional principal = $1"
)
print("=" * 97)
df3
```

```python
df4 = df1.copy()
df4.drop(["Conditional Survival Probability"], axis=1, inplace=True)
df4["Time (years)"] = 0.00
for i in range(0, len(df["Time (years)"])):
    df4["Time (years)"][i] = df1["Time (years)"][i] - 0.5
df4["Expected Accrual Payment"] = (df4["Conditional Default Probability"] * 0.5).round(
    4
)
df4["Financial Discount Factor"] = (1 * np.exp(-r * df4["Time (years)"])).round(4)
df4["APV of Expected Accrual Payment"] = (
    df4["Expected Accrual Payment"] * df4["Financial Discount Factor"]
).round(4)
df4["Expected Accrual Payment"] = (df4["Expected Accrual Payment"]).astype(str) + "s"
df4.loc[len(df4.index)] = [
    "Total",
    "",
    "",
    "",
    np.sum(df4["APV of Expected Accrual Payment"]).round(4),
]
df4["APV of Expected Accrual Payment"] = (
    df4["APV of Expected Accrual Payment"].astype(str) + "s"
)
print("=" * 73)
print("Table 5. Calculation of the actuarial present value of accrual payments")
print("=" * 73)
df4
```
